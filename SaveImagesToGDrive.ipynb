```
import os

DATADIR = '/gdrive/My Drive/Data_test/'

dirs = os.listdir(DATADIR)
dirs.sort()
dirs
```

```
import numpy as np
import pickle

idx = 0
for file in dirs:
  print(file)
  with open(DATADIR+file, 'rb') as f:
    x, y, rest = pickle.load(f)
    for i in range(x.shape[0]):
      
      print('Image_%.6i'%idx)
      folder_idx = int(np.floor((idx/1000)))
      impath = r'/gdrive/My Drive/data/sampled_data_test/images%.2i' % folder_idx
      if not os.path.exists(impath):
        os.makedirs(impath)
      segpath = r'/gdrive/My Drive/data/sampled_data_test/labels%.2i' % folder_idx
      if not os.path.exists(segpath):
        os.makedirs(segpath)
      np.save(impath+'/image_%.6i.npy'%idx,x[i])
      np.save(segpath+'/label_%.6i.npy'%idx,y[i])
      idx += 1
```
